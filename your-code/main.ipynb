#### Plot the correlation matrix using seaborn.

```python
# your code here
corr_matrix = titanic.corr()

# mask duplicates
mask = np.triu(np.ones_like(corr_matrix, dtype = np.bool))

sns.heatmap(corr_matrix, mask = mask, annot = True, fmt="1.2f", annot_kws={'size': 10}, 
            square = True)

# annot = True makes the heatmat not show correctly in the current version of matplotlib
# according to:
# https://datascience.stackexchange.com/questions/57245/seaborn-heatmap-not-displaying-correctly
# and:
# https://stackoverflow.com/questions/56942670/matplotlib-seaborn-first-and-last-row-cut-in-half-of-heatmap-plot
# I kept it beause it's still useful
```
